```python
new_df = df[['title','overview','genres','tagline','vote_average','popularity',]]
```

## Data cleaning

```python
new_df.isnull().sum()
```

```python
new_df=new_df.dropna(subset=['title'])
```

```python

new_df['overview']= new_df['overview'].fillna('')
```

```python
new_df.iloc[0]['genres']
```

```python
import ast
new_df['genres']= new_df['genres'].apply(lambda x : " ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
new_df.head()
```

```python
new_df['tagline']= new_df['tagline'].fillna('')
```

```python
new_df.isnull().sum()
```

```python
new_df['tags']=new_df['overview']+" "+new_df["genres"]+ " " +new_df['tagline']
```
